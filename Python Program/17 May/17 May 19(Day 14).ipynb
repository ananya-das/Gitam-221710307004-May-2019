# Problem2:
  Define a function to add new row  to a csv file
  
### constraint:

### Test Case:

```python
def addRowCSV(filename, rowdata):
    df = readCSV(filename)           #Get data frame from csv
    df.loc[len(df)] = rowdata        # adda a row to the dataframe
    df.to_csv(filename, index = False)              #send the dataframe back to csv
    return df
rowdata = ['1010101', 'statename', 123, 4325, 4565, 9083, 8312, 4564, 8901, 2341,1234]
addRowCSV('DataFiles/Income.csv',rowdata)
```
