# Problem1:
Draw a pie plot based on number of wins for each team


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

class win_pieplot():
    def __init__(self, filename):
        self.df = pd.read_csv(filename)
        self.won=[]
        self.team,self.unique_team=[],[]
        #self.unique_won,self.freq=[],[]
        self.toss=[]
        self.matwin = []
        
    def win_data(self):
        self.team.append(self.df['Team1'].values)
        
        
        for i in range(0,len(self.team[0])):            
            self.toss.append(self.df.iat[i,9])
                
        
        for i in range(0,len(self.team[0])):
            self.matwin.append(self.df.iat[i,10])
            
        
        for i in range(0,len(self.toss)):
            if(self.df.iat[i,9]==self.df.iat[i,10]):
                self.won.append(self.df.iat[i,10])
                
        
        self.unique_won,self.freq = list(np.unique(self.won, return_counts=True))
        
        
        labels = self.unique_won
        number_of_win = self.freq

        fig1, ax1 = plt.subplots()
        ax1.pie(number_of_win, labels=labels, autopct='%1.2f%%',
                shadow=True, startangle=90, radius = 1.8 * 10000)

        ax1.axis('equal') 
        plt.show()     
        
        
final = win_pieplot('Match1.csv')
final.win_data()
```
