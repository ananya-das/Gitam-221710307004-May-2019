```python
claimants=claimants.iloc[:,:].apply(lambda x:x.fillna(x.value_counts().index[0]))
claimants.isnull().sum()
```

```python
import statsmodels.formula.api as smf
logit_model = smf.logit("ATTORNEY ~ CLMSEX + CLMINSUR + SEATBELT + CLMAGE + LOSS", data = claimants).fit()
```

```python
logit_model.summary()
```

```python
import statsmodels.formula.api as smf
logit_model = smf.logit("ATTORNEY ~ CLMSEX + CLMINSUR + SEATBELT +  LOSS", data = claimants).fit()
logit_model.summary()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
claimants.head(1)
```

```python
y_pred1 = logit_model.predict(claimants)
y_pred1
```

```python
claimants["pred_prob"] = y_pred1
claimants.head()
```

```python
claimants["PValue"] = 0
claimants.head()
```

```python
claimants.loc[claimants.pred_prob >= 0.5,"PValue"] = True
claimants.head()
```

```python
confusion_matrix = pd.crosstab(claimants['ATTORNEY'],claimants.PValue)
confusion_matrix
```
